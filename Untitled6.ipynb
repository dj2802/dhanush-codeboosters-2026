```python
# STEP 1: Import Libraries
# ========================
import pandas as pd

print("Libraries imported successfully!")
```

```python
# STEP 2: Load the Dataset
# ============================

df = pd.read_csv('student_performance.csv')

```

```python
# STEP 3: View the dataset
# ============================

print (df)
```

```python
# STEP 4: View the First Few Rows of the Dataset
# ================================================

df.head()
```

```python
# View the LAST 5 rows

df.tail()
```

```python
# STEP 5: Check the Shape
# =========================

print("Shape of dataset:", df.shape)

print("Number of students:", df.shape[0])

print("Number of columns:", df.shape[1])
```

```python
# STEP 6: View All Column Names
# ==============================

print("Columns in the dataset:")
print(df.columns.tolist()
      )
```

```python
# STEP 7: Check Data Types of Each Column
# =========================================

print("Data types:")
print(df.dtypes)
```

```python
# STEP 8: Statistical Summary
# ============================

df.describe()
```

```python
# STEP 9: Check for Missing Values
# =================================

print("Missing values in each column:")
print(df.isnull().sum())

print("\nTotal missing values:", df.isnull().sum().sum())
```

```python
# STEP 10: View a Specific Column
# ================================

print("Student Names:")
print(df['name'])
```

```python
# View unique departments

print("Departments in this dataset:")
print(df['department'].unique())

print("\nHow many students per department:")
print(df['department'].value_counts())
```

```python
# STEP 11: Filter Students by Department
# =========================================

cs_students = df[df['department'] == 'Computer Science']

print(f"Number of Computer Science students: {len(cs_students)}")

print("\nComputer Science students:")
print(cs_students[['name', 'math_score', 'programming_score', 'attendance_percentage']])
```

```python
# Filter students with high attendance (above 90%)

high_attendance = df[df['attendance_percentage'] > 90]

print(f"Students with attendance above 90%: {len(high_attendance)}")
print(high_attendance[['name', 'department', 'attendance_percentage']])
```

```python
df['total_score'] = df['math_score'] + df['science_score'] + df['english_score'] + df['programming_score']

toppers = df[df['total_score']/4 > 90]
print(toppers[['name', 'total_score']])

print("\nNumber of toppers: ", len(toppers))
```

```python
# STEP 12: Find the Top Performing Student
# ==========================================

# First, calculate total score for each student
df['total_score'] = df['math_score'] + df['science_score'] + df['english_score'] + df['programming_score']

# Find the student with the highest total score
top_student = df.loc[df['total_score'].idxmax()]


print("Top Student:")
print(f"Name: {top_student['name']}")
print(f"Department: {top_student['department']}")
print(f"Total Score: {top_student['total_score']}")
print(f"Attendance: {top_student['attendance_percentage']}%")
```

```python
dept_avg =df.groupby('department')['total_score'].mean()
print("average total score per department(Desending order)")
print(dept_avg.sort_values(ascending=False))
```

```python
# Calculate average score per department

dept_avg = df.groupby('department')['total_score'].mean().round(2)

print("Average Total Score by Department:")
print(dept_avg)
```

```python
# STEP 13: Sort Students by Total Score (Highest to Lowest)
# ==========================================================

df_sorted = df.sort_values(by='total_score', ascending=False)

print("Top 10 Students by Total Score:")
print(df_sorted[['name', 'department', 'total_score', 'attendance_percentage']].head(13))
```

```python
# MINI PROJECT
# Student Data Explorer
# =====================================

print("="*60)
print("STUDENT PERFORMANCE EXPLORER REPORT")
print("="*60)

# --- Section 1: Dataset Overview ---
print("\n[1] DATASET OVERVIEW")
print("-"*40)
print(f"Total Students: {df.shape[0]}")
print(f"Total Columns: {df.shape[1]}")
print(f"Departments: {df['department'].nunique()}")

# --- Section 2: Department-wise Analysis ---
print("\n[2] DEPARTMENT-WISE STUDENT COUNT")
print("-"*40)
print(df['department'].value_counts())

# --- Section 3: Score Analysis ---
print("\n[3] SCORE ANALYSIS")
print("-"*40)
print(f"Highest Math Score: {df['math_score'].max()}")
print(f"Lowest Math Score: {df['math_score'].min()}")
print(f"Average Math Score: {df['math_score'].mean():.2f}")
print(f"Average Programming Score: {df['programming_score'].mean():.2f}")
print(f"Average Attendance: {df['attendance_percentage'].mean():.2f}%")

# --- Section 4: Gender-wise Analysis ---
print("\n[4] GENDER-WISE AVERAGE SCORES")
print("-"*40)
gender_scores = df.groupby('gender')['total_score'].mean().round(2)
print(gender_scores)

# --- Section 5: Top Performers ---
print("\n[5] TOP 5 PERFORMERS")
print("-"*40)
top5 = df.sort_values('total_score', ascending=False).head(5)
for index, row in top5.iterrows():
    print(f"  {row['name']:<20} | {row['department']:<18} | Score: {row['total_score']}")


# --- Section 6: Students Who Need Attention ---
print("\n[6] STUDENTS WITH LOW ATTENDANCE (Below 75%)")
print("-"*40)
low_att = df[df['attendance_percentage'] < 75][['name', 'department', 'attendance_percentage']]
if len(low_att) > 0:
    print(low_att.to_string(index=False))

else:
    print("All students have attendance above 75%")

print("\n" + "="*60)
print("Report Complete")
print("="*60)
```
